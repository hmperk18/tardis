## Construct a table of energy packets that can be filtered by the element from which it came from

The `SDECPlotter` object holds a dataframe with the energy packets information along with the element and ion species involved in the packet's last interaction, as that information is used to color the emission and absorption curves seen above. 

For this example work, I will focus on the real packets, but I could use data stored in the `vpacket_tracker` for the virtual packets.

```python
packets_mode = 'real'
plotter.data[packets_mode].packets_df_line_interaction
```

This table was constructed from data sorted in the simulation's `transport_state` for the packets that experience an interaction

```python
# obtaining the basic packet information, as prepared in SDECData.from_simulation

ts = sim.transport.transport_state
em_mask = ts.emitted_packet_mask
packets_df = pd.DataFrame({
    "nus": ts.packet_collection.output_nus[em_mask],
    "lambdas": (ts.packet_collection.output_nus[em_mask]*u.Hz).to(u.AA, equivalencies=u.spectral()),
    "energies": ts.packet_collection.output_energies[em_mask],
    "last_interaction_type": ts.last_interaction_type[em_mask],
    "last_line_interaction_out_id": ts.last_line_interaction_out_id[em_mask],
    "last_line_interaction_in_id": ts.last_line_interaction_in_id[em_mask],
    "last_line_interaction_in_nu": ts.last_interaction_in_nu[em_mask],
    "last_interaction_in_r": ts.last_interaction_in_r[em_mask]
    }
)

# remove all 'no interactions'
int_mask = (packets_df["last_interaction_type"] > -1) & (packets_df["last_line_interaction_in_id"] > -1)
packets_df = packets_df[int_mask]

# get lines information
lines_df = sim.plasma.atomic_data.lines.reset_index().set_index("line_id")

# for the packets that were emitted, get the atom and species from the last line interaction
packets_df["last_line_interaction_atom"] = lines_df.iloc[packets_df["last_line_interaction_out_id"]]["atomic_number"].values
packets_df["last_line_interaction_ion"] = lines_df.iloc[packets_df["last_line_interaction_out_id"]]["ion_number"].values

# per tardis convention, the species is Z*100 + ion_number
packets_df["last_line_interaction_species"] = packets_df["last_line_interaction_atom"]*100 + packets_df["last_line_interaction_ion"]
```

```python
packets_df
```

The table above thus has the atomic number of the line the packet last interacted with, or the element it originated from, and can be filtered by that column.

## Plot the Table

To plot the table, I will construct a histogram showing the distribution of packets from the simulation in energy for each of the elements involved. The `SDECPlotter` nicely associates the relevant elements with `matplotlib` colors, so to make use of the functions within there, I will clear the axes and fill with my histograms.

```python
# using the basis of the generate_plot_mpl and associated functions of sdec_plot.py

figsize = (12, 7)
ax = plotter.generate_plot_mpl()
ax.cla() # clear the data, but keep the colorbar

# for each element, plot the distribution of associated packet energies
for species_counter, identifier in enumerate(plotter.species):
    mask = packets_df['last_line_interaction_atom'] == identifier
    energies = packets_df[mask]['energies'] # <-- example of filtering the table by origin element

    ax.hist(energies, bins=100, histtype='step', color=plotter._color_list[species_counter])
    

ax.set_xlabel('Energy [erg]', fontsize=12)
ax.set_ylabel('Number of Packets', fontsize=12)
ax.set_title('Distribution of Packet Energies per Originating Atomic Species')
plt.show()
```
