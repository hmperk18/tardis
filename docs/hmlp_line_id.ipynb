## Construct a table of energy packets that can be filtered by the element from which it came from

The `SDECPlotter` object holds a dataframe with the energy packets information along with the element and ion species involved in the packet's last interaction, as that information is used to color the emission and absorption curves seen above. 

For this example work, I will focus on the real packets, but I could use data stored in the `vpacket_tracker` for the virtual packets.

```python
packets_mode = 'real'
plotter.data[packets_mode].packets_df_line_interaction
```

This table was constructed from data sorted in the simulation's `transport_state` for the packets that experience an interaction

```python
# obtaining the basic packet information, as prepared in SDECData.from_simulation

ts = sim.transport.transport_state
em_mask = ts.emitted_packet_mask
packets_df = pd.DataFrame({
    "nus": ts.packet_collection.output_nus[em_mask],
    "lambdas": (ts.packet_collection.output_nus[em_mask]*u.Hz).to(u.AA, equivalencies=u.spectral()),
    "energies": ts.packet_collection.output_energies[em_mask],
    "last_interaction_type": ts.last_interaction_type[em_mask],
    "last_line_interaction_out_id": ts.last_line_interaction_out_id[em_mask],
    "last_line_interaction_in_id": ts.last_line_interaction_in_id[em_mask],
    "last_line_interaction_in_nu": ts.last_interaction_in_nu[em_mask],
    "last_interaction_in_r": ts.last_interaction_in_r[em_mask]
    }
)

# remove all 'no interactions'
int_mask = (packets_df["last_interaction_type"] > -1) & (packets_df["last_line_interaction_in_id"] > -1)
packets_df = packets_df[int_mask]

# get lines information
lines_df = sim.plasma.atomic_data.lines.reset_index().set_index("line_id")

# for the packets that were emitted, get the atom and species from the last line interaction
packets_df["last_line_interaction_atom"] = lines_df.iloc[packets_df["last_line_interaction_out_id"]]["atomic_number"].values
packets_df["last_line_interaction_ion"] = lines_df.iloc[packets_df["last_line_interaction_out_id"]]["ion_number"].values

# per tardis convention, the species is Z*100 + ion_number
packets_df["last_line_interaction_species"] = packets_df["last_line_interaction_atom"]*100 + packets_df["last_line_interaction_ion"]
```

```python
packets_df
```

The table above thus has the atomic number of the line the packet last interacted with, or the element it originated from, and can be filtered by that column.

## Plot the Table

To plot the table, I will construct a histogram showing the distribution of packets from the simulation in wavelength for each of the elements. This allows us to see how many packets from a given element is contributing to each wavelength The `SDECPlotter` nicely associates the relevant elements with `matplotlib` colors, so to make use of the functions within there, I will clear the axes and fill with my histograms.

If this were to be a real plotting script in TARDIS, I would build a similar data object and utilize the plotting functions without first needing to empty the plot generated from the `SDECPlotter`.

```python
# using the basis of the generate_plot_mpl and associated functions of sdec_plot.py
figsize = (12, 7)
ax = plotter.generate_plot_mpl()
ax.cla() # clear the data, but keep the colorbar

# plot full distr
    # save the bin widths to use for each element
_, bins, _ = ax.hist(packets_df['lambdas'], bins=250, color='gray', label='All Packets', alpha=0.5, zorder=-len(packets_df['lambdas']))

# for each element, plot the distribution of associated packet energies
for species_counter, identifier in enumerate(plotter.species):
    mask = packets_df['last_line_interaction_atom'] == identifier
    wavs = packets_df[mask]['lambdas'] # <-- example of filtering the table by origin element

    ax.hist(wavs, bins=bins, color=plotter._color_list[species_counter], zorder=-len(wavs))

ax.set_xlabel(r'Wavelength [\AA]', fontsize=12)
ax.set_ylabel('Number of Packets', fontsize=12)
ax.set_title('Distribution of Packet Energies per Originating Atomic Species')
ax.set_xlim(0, 1e4) # cut off region with few packets
plt.show()
```

# Preliminary Line Identification

From the distribution above, it is clear that there are some interesting wavelength associated with different elements.  I am going to focus on silicon first.

```python
figsize = (12, 7)
fig, ax = plt.subplots(1,1, figsize=figsize)

atomic_number = 14 
mask = packets_df['last_line_interaction_atom'] == atomic_number
atom_packets = packets_df[mask]

_, bins, _ = ax.hist(atom_packets['lambdas'],bins=250, color='gray', zorder = -len(atom_packets['lambdas']),label='All Si Packets')

# loop through each ionization state
for species_counter, identifier in enumerate(atom_packets['last_line_interaction_species'].unique()):
    mask = atom_packets['last_line_interaction_species'] == identifier
    ion = (identifier % 100)
    wavs = atom_packets[mask]['lambdas']
    
    label = f'{int_to_roman(ion+1)}' # have to add 1 because neutral (0) is I (one)
    h = ax.hist(wavs, bins=bins, color=f'C{species_counter}', zorder = -len(wavs), label=f'{int_to_roman(ion+1)}')

ax.set_xlabel(r'Wavelength [\AA]', fontsize=12)
ax.set_ylabel('Number of Packets', fontsize=12)
ax.set_title('Distribution of Packet Energies per Ionization State of Silicon')
ax.legend()
ax.set_xlim(0, 1e4) # cut off region with few packets
plt.show()
```

Even within just silicon there are interesting features associated with each ionzation state. Let's find the locations of these peaks. As a first pass, I am going to follow [this stack overflow example](https://stackoverflow.com/questions/67145403/is-there-a-way-i-can-find-the-range-of-local-maxima-of-histogram).

Selecting 2 peaks per ionization found in this way, I have used the locations to then place line markers at the top of the plot to indicate the found peaks.

```python
figsize = (12, 7)
fig, ax = plt.subplots(1,1, figsize=figsize)

atomic_number = 8 
mask = packets_df['last_line_interaction_atom'] == atomic_number
atom_packets = packets_df[mask]

_, bins, _ = ax.hist(atom_packets['lambdas'],bins=250, color='gray', zorder = -len(atom_packets['lambdas']), label='All Si Packets')

# loop through each ionization state
for species_counter, identifier in enumerate(atom_packets['last_line_interaction_species'].unique()):
    mask = atom_packets['last_line_interaction_species'] == identifier
    ion = (identifier % 100)
    wavs = atom_packets[mask]['lambdas']
  
    # peak finding - based on stack overflow answer
            # https://stackoverflow.com/questions/67145403/is-there-a-way-i-can-find-the-range-of-local-maxima-of-histogram
    
    label = f'{int_to_roman(ion+1)}' # have to add 1 because neutral (0) is I (one)
    ax.hist(wavs, bins=bins, color=f'C{species_counter}', zorder = -len(wavs), label=f'{int_to_roman(ion+1)}')
    
    # normalize the packet counts
    h = ax.hist(wavs, bins=bins, color=f'C{species_counter}', zorder = -len(wavs), alpha=0, density=True)

    density = h[0]
    values = h[1][:-1] + np.diff(h[1])[0] / 2 # get the center of the bins
    norm_density = density / density.sum()
    peaks = find_peaks(norm_density, prominence=0.1)[0] # set a prominence so that they are robust selection of the peaks

    # select the first 2 peak for each state 
    for peak in peaks[:2]:
        plt.axvline(values[peak], ymin = 0.95, ymax = 1, color=f'C{species_counter}')

ax.set_xlabel(r'Wavelength [\AA]', fontsize=12)
ax.set_ylabel('Number of Packets', fontsize=12)
ax.set_title('Distribution of Packet Energies per Ionization State of Silicon')
ax.legend()
ax.set_xlim(0, 1e4) # cut off region with few packets
plt.show()
```

Now run through each element.

I scaled the length of the marker lines, which reveals locations of overalapping features.

```python
# using the basis of the generate_lot_mpl and associated functions of sdec_plot.py

figsize = (12, 7)
ax = plotter.generate_plot_mpl()
ax.cla()

# hidden full distribution to save the bin widths
_, bins, _ = ax.hist(packets_df['lambdas'],bins=250, color='gray', zorder = -len(packets_df['lambdas']))

ls = ["-",":","--","-."]  # use a different line style for each ionization state 

# loop through each element
for atom_counter, el_identifier in enumerate(plotter.species):
    mask = packets_df['last_line_interaction_atom'] == el_identifier
    atom_packets = packets_df[mask]
    wavs = atom_packets['lambdas']
    _, bins, _ = ax.hist(wavs, bins=bins, color=plotter._color_list[atom_counter], zorder = -len(wavs))

    # loop through each ionization state
    for species_counter, identifier in enumerate(atom_packets['last_line_interaction_species'].unique()):
        mask = atom_packets['last_line_interaction_species'] == identifier
        ion = (identifier % 100)
        wavs = atom_packets[mask]['lambdas']
        
        h = ax.hist(wavs, bins=bins, zorder = -len(wavs), alpha=0, density=True)

        # peak finding - based on stack overflow answer
            # https://stackoverflow.com/questions/67145403/is-there-a-way-i-can-find-the-range-of-local-maxima-of-histogram
        density = h[0]
        values = h[1][:-1] + np.diff(h[1])[0] / 2
        norm_density = density / density.sum()
        peaks = find_peaks(norm_density, prominence=0.1)[0]

        # select the first 2 peak for each state
            # scale line by the element and sort so longer line is behind the shorter lines
        for peak in peaks[:2]:
            length = 0.1 + (0.005*el_identifier) 
            plt.axvline(values[peak], ymin = 1-length, ymax = 1, color=plotter._color_list[atom_counter], linestyle=ls[ion], linewidth=2, zorder = 10 - length)
        
# create hidden lines for cleaner legend
for ion in (packets_df['last_line_interaction_species'] % 100).unique():
    label = f'{int_to_roman(ion+1)}'
    plt.axvline(-1000, color='gray', label=label, linestyle=ls[ion])        
        
ax.set_xlabel(r'Wavelength [\AA]', fontsize=12)
ax.set_ylabel('Number of Packets', fontsize=12)
ax.set_title('Distribution of Packet Energies per Ionization State of Silicon')
ax.legend()
ax.set_ylim(0, 2700)
ax.set_xlim(0, 1e4) # cut off region with few packets
plt.show()
```
